### Analizar el comportamiento de compra de los clientes:

#### Calcula el valor promedio de los pedidos realizados por cada cliente.

```python
valor_promedio_cliente = pd.read_sql_query(
    '''
    SELECT c.customer_unique_id, ROUND(AVG(price),2) as precio_promedio_pedido
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    INNER JOIN order_items oi ON
    oi.order_id=o.order_id
    GROUP BY c.customer_unique_id,o.order_id
    ORDER BY precio_promedio_pedido DESC
    ''',
    conn
)

#orders es tabla intermedia que unifica order_id y customer_id
#order_items tiene el order_id y el price + freight_value

valor_promedio_cliente.to_csv('valor_promedio_cliente')

valor_promedio_cliente
```

#### Identifica a los clientes con los pedidos de mayor valor.

```python
clientes_pedidos_caros = pd.read_sql_query(
    '''
    SELECT c.customer_unique_id, SUM(price) as suma_precio_pedidos
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    INNER JOIN order_items oi ON
    oi.order_id=o.order_id
    GROUP BY c.customer_unique_id
    ORDER BY SUM(price) DESC
    LIMIT 10
    ''',
    conn
)

#orders es tabla intermedia que unifica order_id y customer_id
#order_items tiene el order_id y el price + freight_value

clientes_pedidos_caros.to_csv('clientes_pedidos_caros')

clientes_pedidos_caros
```

#### Calcular la frecuencia promedio de compra por cliente, es decir, la cantidad de pedidos realizados en un período de tiempo determinado.

```python
fechas_operaciones = pd.read_sql_query(
    '''
    SELECT julianday(max(o.order_purchase_timestamp)) as fecha_max,
           julianday(min(o.order_purchase_timestamp)) as fecha_min,
           julianday(max(o.order_purchase_timestamp)) - julianday(min(o.order_purchase_timestamp)) as tiempo_operacion_empresa
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    INNER JOIN order_items oi ON
    oi.order_id=o.order_id
    ''',
    conn
)

#orders es tabla intermedia que unifica order_id y customer_id
#orders tiene tiempo de compra order_purchase_timestamp
#order_items tiene el order_id y el price + freight_value

fechas_operaciones.to_csv('fechas_operaciones')

fechas_operaciones
```

```python
cliente_freq_compra = pd.read_sql_query(
    '''
    SELECT  c.customer_unique_id,
            COUNT(o.order_id) as cantidad_compras,
            COUNT(o.order_id)/728.49419 as freq_compra,
            (COUNT(o.order_id)/728.49419)*100 as freq_compra_x100
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    INNER JOIN order_items oi ON
    oi.order_id=o.order_id
    GROUP BY c.customer_unique_id
    ORDER BY freq_compra DESC
    ''',
    conn
)

#orders es tabla intermedia que unifica order_id y customer_id
#orders tiene tiempo de compra order_purchase_timestamp
#order_items tiene el order_id y el price + freight_value
cliente_freq_compra.to_csv('cliente_freq_compra')

cliente_freq_compra
```

#### Generar un informe que muestre los clientes más frecuentes y su promedio de frecuencia de compra.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
fechas_operaciones = pd.read_sql_query(
    '''
    SELECT julianday(max(o.order_purchase_timestamp)) as fecha_max,
           julianday(min(o.order_purchase_timestamp)) as fecha_min,
           julianday(max(o.order_purchase_timestamp)) - julianday(min(o.order_purchase_timestamp)) as tiempo_operacion_empresa
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    INNER JOIN order_items oi ON
    oi.order_id=o.order_id
    ''',
    conn
)

#orders es tabla intermedia que unifica order_id y customer_id
#orders tiene tiempo de compra order_purchase_timestamp
#order_items tiene el order_id y el price + freight_value

fechas_operaciones.to_csv('fechas_operaciones')

fechas_operaciones
```

```python
cliente_freq_compra = pd.read_sql_query(
    '''
    SELECT  c.customer_unique_id,
            COUNT(o.order_id) as cantidad_compras,
            COUNT(o.order_id)/728.49419 as freq_compra,
            ROUND((COUNT(o.order_id)/728.49419)*100,2) as freq_compra_x100
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    INNER JOIN order_items oi ON
    oi.order_id=o.order_id
    GROUP BY c.customer_unique_id
    ORDER BY freq_compra DESC
    LIMIT 10
    ''',
    conn
)

#orders es tabla intermedia que unifica order_id y customer_id
#orders tiene tiempo de compra order_purchase_timestamp
#order_items tiene el order_id y el price + freight_value

cliente_freq_compra.to_csv('cliente_freq_compra_limit10')

cliente_freq_compra
```

```python
plt.figure(layout='constrained')
#plt.bar(tiempo_entrega_rapido['customer_id'],tiempo_entrega_rapido['horas_transcurridas'])
fig, ax = plt.subplots()
c = ['green', 'orange', 'orange', 'orange', 'orange','orange', 'orange', 'orange','red','red']
bars = ax.bar(cliente_freq_compra['customer_unique_id'],cliente_freq_compra['freq_compra_x100'],color=c)

ax.bar_label(bars)

plt.xlabel('Clientes')
plt.ylabel('Compras cada 100 días')
plt.title("Top 10 Clientes por Frecuencia de compra cada 100 días")
plt.legend()
plt.tick_params(axis='x', labelrotation=90)
ax.set_xticks([])

plt.savefig('freq_compra_cada_100_dias.png')
```

En el gráfico de arriba podemos ver e**l Top 10 de Clientes con mayor frecuencia de compra**. Dada que la frecuencia de compra siempre es inferior a una compra por día, se realizó una operación matemática para calcular **la tasa de compra cada 100 días**, lo cual consideramos facilita la interpretación de los datos. De esta forma lo que vemos es que el **primer cliente (color verde), cada 100 días realiza poco más de 3 compras/pedidos**. Del **segundo al octavo cliente (color naranja), cada 100 días hacen 2 compras**. **Los últimos dos clientes (color rojo) de nuestro top 10 hacen poco menos de dos pedidos cada 100 días**.

Para calcular el umbral de tiempo en base al cual calcular esta tasa, en primer lugar se calculó la diferencia entre la primera y la última compra de la empresa, dando un total de casi 730 días (728.49419 exactamente). En base a ello, se calculó la cantidad de pedidos realizados por cada cliente y luego se lo dividió por este total de días, dando la frecuencia en términos proporcionales. Finalmente se multiplicó este número por 100 para transformar esa proporción en la ya mencionada tasa de compra cada 100 días.

```python
tiempo_promedio_entrega = pd.read_sql_query(
    '''
    SELECT order_status,AVG(((strftime("%s", o.order_delivered_customer_date) - strftime("%s", o.order_purchase_timestamp))/60)/60) AS horas_promedio_entrega,
    AVG((((strftime("%s", o.order_delivered_customer_date) - strftime("%s", o.order_purchase_timestamp))/60)/60)/24) AS dias_promedio_entrega
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    GROUP BY order_status
    HAVING order_status='delivered'
    AND horas_promedio_entrega IS NOT NULL
    ''',
    conn
)

#customers tiene customer_id
#orders tiene order_status (8 estados), tiempo compra order_purchase_timestamp y order_delivered_customer_date

tiempo_promedio_entrega.to_csv('tiempo_promedio_entrega')

tiempo_promedio_entrega
```

#### Genera un informe que muestre los clientes con los tiempos de entrega más rápidos y más lentos.

```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tiempo_entrega_rapido = pd.read_sql_query(
    '''
    SELECT  o.customer_id,
            AVG(((strftime("%s", o.order_delivered_customer_date) - strftime("%s", o.order_purchase_timestamp))/60)/60) AS horas_transcurridas,
            AVG((((strftime("%s", o.order_delivered_customer_date) - strftime("%s", o.order_purchase_timestamp))/60)/60)/24) AS dias_transcurridos
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    WHERE order_status='delivered'
    GROUP BY o.customer_id
    HAVING dias_transcurridos IS NOT NULL
    ORDER BY horas_transcurridas ASC
    LIMIT 10
    ''',
    conn
)

#customers tiene customer_id
#orders tiene order_status (8 estados), tiempo compra order_purchase_timestamp y order_delivered_customer_date

tiempo_entrega_rapido.to_csv('tiempo_entrega_rapido')

tiempo_entrega_rapido
```

```python
tiempo_entrega_lento = pd.read_sql_query(
    '''
    SELECT  o.customer_id,
            AVG(((strftime("%s", o.order_delivered_customer_date) - strftime("%s", o.order_purchase_timestamp))/60)/60) AS horas_transcurridas,
            AVG((((strftime("%s", o.order_delivered_customer_date) - strftime("%s", o.order_purchase_timestamp))/60)/60)/24) AS dias_transcurridos
    FROM customers c
    INNER JOIN orders o ON
    o.customer_id=c.customer_id
    WHERE order_status='delivered'
    GROUP BY o.customer_id
    HAVING dias_transcurridos IS NOT NULL
    ORDER BY horas_transcurridas DESC
    LIMIT 10
    ''',
    conn
)

#customers tiene customer_id
#orders tiene order_status (8 estados), tiempo compra order_purchase_timestamp y order_delivered_customer_date

tiempo_entrega_lento.to_csv('tiempo_entrega_lento')

tiempo_entrega_lento
```

```python
plt.figure(layout='constrained')
#plt.bar(tiempo_entrega_rapido['customer_id'],tiempo_entrega_rapido['horas_transcurridas'])
fig, ax = plt.subplots()
bars = ax.bar(tiempo_entrega_rapido['customer_id'],tiempo_entrega_rapido['horas_transcurridas'],color = 'green')

ax.bar_label(bars)

plt.xlabel('Clientes')
plt.ylabel('Horas')
plt.title("Top 10 Entregas más rápidas En Horas Transcurridas desde la compra")
plt.legend()
plt.tick_params(axis='x', labelrotation=90)
ax.set_xticks([])

plt.savefig('tiempo_entrega_rapido.png')
```

En el gráfico de arriba podemos ver el **Top 10 de los tiempos de entrega más rápidos**, **calculado en horas transcurridas desde el momento de la compra**. Cada columna representa a un cliente único, pero el eje X fue ocultado por cuestiones de legibilidad del gráfico. En principio, podemos observar que **todas las entregas fueron realizadas en un período inferior a un día**, es decir a 24 horas desde el momento de compra. A su vez, **el cliente que más rápido recibió su pedido fue en un período de medio día, es decir 12 horas**.

```python
plt.figure(layout='constrained')
#plt.bar(tiempo_entrega_rapido['customer_id'],tiempo_entrega_rapido['horas_transcurridas'])
fig, ax = plt.subplots()
bars = ax.bar(tiempo_entrega_lento['customer_id'],tiempo_entrega_lento['dias_transcurridos'],color = 'orange')

ax.bar_label(bars)

plt.xlabel('Clientes')
plt.ylabel('Días')
plt.title("Top 10 Entregas más lentas En Días Transcurridas desde la compra")
plt.legend()
plt.tick_params(axis='x', labelrotation=90)
ax.set_xticks([])

plt.savefig('tiempo_entrega_lento.png')
```

Analizando el gráfico de arriba, lo que tenemos es el **Top 10 de entregas más lentas, en este caso calculado en días transcurridos desde la compra** para facilitar la lectura e interpretación de la información. **En todos estos casos, el tiempo de entrega demoró más de medio año**, es decir 182,5 días. Es interesante remarcar que, mientras en el gráfico de pedidos más rápidos ninguno llegó a demorar más de un día, en este caso los pedidos demoraron como mínimo medio año para llegar a su cliente. A su vez, **teniendo en cuenta que el tiempo promedio de entrega es de 12 días aproximadamente, las entregas más lentos son al menos 15 veces más lentas que dicho promedio**.

## Analizar el rendimiento de los métodos de pago utilizados por los clientes:

```python
# Esta función tiene el propósito de evitar la notación científica frente a números muy grandes en los gráficos

def reformat_large_tick_values(tick_val, pos):
    """
    Turns large tick values (in the billions, millions and thousands) such as 4500 into 4.5K and also appropriately turns 4000 into 4K (no zero after the decimal).
    """
    if tick_val >= 1000000000:
        val = round(tick_val/1000000000, 1)
        new_tick_format = '{:}B'.format(val)
    elif tick_val >= 1000000:
        val = round(tick_val/1000000, 1)
        new_tick_format = '{:}M'.format(val)
    elif tick_val >= 1000:
        val = round(tick_val/1000, 1)
        new_tick_format = '{:}K'.format(val)
    elif tick_val < 1000:
        new_tick_format = round(tick_val, 1)
    else:
        new_tick_format = tick_val

    # make new_tick_format into a string value
    new_tick_format = str(new_tick_format)

    # code below will keep 4.5M as is but change values such as 4.0M to 4M since that zero after the decimal isn't needed
    index_of_decimal = new_tick_format.find(".")

    if index_of_decimal != -1:
        value_after_decimal = new_tick_format[index_of_decimal+1]
        if value_after_decimal == "0":
            # remove the 0 after the decimal point since it's not needed
            new_tick_format = new_tick_format[0:index_of_decimal] + new_tick_format[index_of_decimal+2:]

    return new_tick_format

plt.figure(layout='constrained')
fig, ax = plt.subplots()
ax.yaxis.set_major_formatter(tick.FuncFormatter(reformat_large_tick_values));

bars = ax.bar(metodo_de_pago['payment_type'],metodo_de_pago['total_dinero'],color=palette_color)

#ax.bar_label(bars)

plt.xlabel('Métodos de pagos')
plt.ylabel('Dinero Operado en Millones')
plt.title("Métodos de pagos según cantidad de dinero operado")
plt.legend()
plt.tick_params(axis='x')

plt.savefig('metodo_de_pago_dinero.png')
```

```python
palette_color = sns.color_palette("pastel")

plt.pie(metodo_de_pago['total_dinero'][0:-1], labels=metodo_de_pago['payment_type'][0:-1], colors=palette_color, autopct='%.0f%%')

plt.title("Métodos de pagos según porcentaje de dinero operado")
plt.savefig('metodo_de_pago_dinero2.png')
plt.show()
```

Veamos qué sucede cuando analizamos **los gráficos de métodos de pago pero por cantidad de dinero operado**. Nuevamente, el gráfico de barras trabaja con los números absolutos, mientras que el gráfico de tortas lo hace con porcentajes. Vemos aquí nuevamente **un patrón similar siendo cómoda la tarjeta de crédito el método a través del cual más dinero se operó, totabilizando el 78% del dinero operado (más de 12 millones de pesos)**. En **segundo lugar, el boleto tiene porcentajes casi idénticos al de cantidad de operaciones, 18%** y un monto operado de cerca de $3 millones de pesos. No así **el voucher que, aunque representaba el 6% de las operaciones, aquí vemos que sólo representa el 2% del dinero total operado. Esto indica que el voucher moviliza en promedio menos dinero por operación que el resto de los métodos de pago**. Por su lado, la tarjeta de débito nuevamente representa el 1% del total, en este caso del monto de dinero operado.
